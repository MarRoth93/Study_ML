## 1. The Apples and Bananas Problem: A Gateway to Linear Algebra

### 🍎🍌 The Classic Setup

Imagine you're at a fruit market with these clues:

- **Purchase 1**: 2 apples + 3 bananas = €8  
- **Purchase 2**: 10 apples + 1 banana = €13

**The Question**: What's the price of 1 apple and 1 banana?

### 📝 Mathematical Formulation

Let's denote:
- `a` = price of 1 apple (€)
- `b` = price of 1 banana (€)

Our **system of linear equations**:

$$\begin{align}
2a + 3b &= 8 \quad \text{(Purchase 1)} \\
10a + 1b &= 13 \quad \text{(Purchase 2)}
\end{align}$$

### 🔍 Why This Problem is Special

This simple scenario contains **all the fundamental concepts** of linear algebra:

| **Concept** | **In Our Problem** | **General Principle** |
|-------------|-------------------|---------------------|
| **Variables** | apple price `a`, banana price `b` | Unknown quantities to find |
| **Coefficients** | {2,3}, {10,1} | How variables combine |
| **Constants** | 8, 13 | Target outcomes |
| **System** | Two equations, two unknowns | Multiple constraints |

### 🎯 Three Ways to View This Problem

#### 1. **🔧 Algebraic View** (Traditional)
Solve by substitution or elimination:
```
From equation 1: a = (8 - 3b)/2
Substitute into equation 2: 10(8 - 3b)/2 + b = 13
```

#### 2. **📐 Geometric View** 
Each equation represents a **line** in the (a,b) plane:
- Line 1: All (a,b) pairs satisfying 2a + 3b = 8
- Line 2: All (a,b) pairs satisfying 10a + b = 13
- **Solution**: Where the lines intersect!

#### 3. **🎨 Vector View** (Our Focus)
Rewrite as: "Find weights to combine direction vectors"
```
a × [2, 10] + b × [3, 1] = [8, 13]
```

### 💡 The Big Insight

The **vector view** reveals that we're asking:
> "How can we combine the direction vectors [2,10] and [3,1] to reach the target point [8,13]?"

This perspective transforms equation-solving into **geometric intuition** and scales beautifully to higher dimensions, making it the foundation for:
- Machine learning algorithms
- Computer graphics transformations  
- Data analysis techniques
- Scientific computing methods

Let's solve this step by step! 🚀


````python
import numpy as np

# === COMPREHENSIVE SOLUTION ANALYSIS ===
print("🔍 SOLVING THE APPLES AND BANANAS SYSTEM")
print("="*60)

# === STEP 1: Set Up the Matrix-Vector Equation ===
print("📊 STEP 1: Matrix-Vector Formulation")
print("="*40)

# Define the coefficient matrix A and target vector b
A = np.array([[2, 3],      # Coefficients for equation 1: 2a + 3b = 8
              [10, 1]])    # Coefficients for equation 2: 10a + 1b = 13

b = np.array([8, 13])      # Right-hand side values

print(f"📝 System of equations:")
print(f"   2a + 3b = 8")
print(f"   10a + 1b = 13")

print(f"\n🔢 Matrix form: A × x = b")
print(f"   Matrix A (coefficients):\n{A}")
print(f"   Vector b (targets): {b}")
print(f"   Vector x (unknowns): [a, b] ← This is what we want to find!")

# === STEP 2: Analyze the System Properties ===
print(f"\n📊 STEP 2: System Analysis")
print("="*40)

# Check if system is well-posed
det_A = np.linalg.det(A)
cond_A = np.linalg.cond(A)

print(f"🔍 System properties:")
print(f"   Determinant of A: {det_A:.6f}")
print(f"   Condition number: {cond_A:.6f}")

if abs(det_A) > 1e-10:
    print(f"   ✅ det(A) ≠ 0 → System has unique solution")
else:
    print(f"   ❌ det(A) ≈ 0 → System is singular (no unique solution)")

if cond_A < 100:
    print(f"   ✅ Well-conditioned → Numerically stable solution")
else:
    print(f"   ⚠️  Ill-conditioned → Solution may be sensitive to errors")

# === STEP 3: Solve Using Multiple Methods ===
print(f"\n🎯 STEP 3: Solution Methods")
print("="*40)

# Method 1: NumPy's linear solver
print(f"📊 Method 1: NumPy Linear Solver")
solution = np.linalg.solve(A, b)
apple_price, banana_price = solution
print(f"   Solution: [a, b] = {solution}")
print(f"   🍎 Apple price: €{apple_price:.4f}")
print(f"   🍌 Banana price: €{banana_price:.4f}")

# Method 2: Matrix inversion (A^-1 × b)
print(f"\n📊 Method 2: Matrix Inversion")
A_inv = np.linalg.inv(A)
solution_inv = A_inv @ b
print(f"   A⁻¹ = \n{A_inv}")
print(f"   Solution = A⁻¹ × b = {solution_inv}")
print(f"   ✅ Match with Method 1: {np.allclose(solution, solution_inv)}")

# Method 3: Cramer's rule
print(f"\n📊 Method 3: Cramer's Rule")
A_a = A.copy()
A_a[:, 0] = b  # Replace first column with b
det_A_a = np.linalg.det(A_a)

A_b = A.copy()  
A_b[:, 1] = b  # Replace second column with b
det_A_b = np.linalg.det(A_b)

apple_cramer = det_A_a / det_A
banana_cramer = det_A_b / det_A

print(f"   Apple price = det(A_a)/det(A) = {det_A_a:.6f}/{det_A:.6f} = {apple_cramer:.4f}")
print(f"   Banana price = det(A_b)/det(A) = {det_A_b:.6f}/{det_A:.6f} = {banana_cramer:.4f}")
print(f"   ✅ Match with Method 1: {np.allclose([apple_cramer, banana_cramer], solution)}")

# === STEP 4: Verification ===
print(f"\n✅ STEP 4: Solution Verification")
print("="*40)

# Check that our solution satisfies both equations
check1 = 2 * apple_price + 3 * banana_price
check2 = 10 * apple_price + 1 * banana_price

print(f"🔍 Substitution check:")
print(f"   Equation 1: 2({apple_price:.4f}) + 3({banana_price:.4f}) = {check1:.6f} ≈ 8")
print(f"   Equation 2: 10({apple_price:.4f}) + 1({banana_price:.4f}) = {check2:.6f} ≈ 13")

error1 = abs(check1 - 8)
error2 = abs(check2 - 13)
print(f"   Error in equation 1: {error1:.2e}")
print(f"   Error in equation 2: {error2:.2e}")

if error1 < 1e-10 and error2 < 1e-10:
    print(f"   ✅ Solution is correct!")
else:
    print(f"   ❌ Solution has errors")

# === STEP 5: Economic Interpretation ===
print(f"\n💰 STEP 5: Economic Interpretation")
print("="*40)

print(f"🍎🍌 Final Answer:")
print(f"   • One apple costs €{apple_price:.2f}")
print(f"   • One banana costs €{banana_price:.2f}")

# Verify with original purchases
purchase1_cost = 2 * apple_price + 3 * banana_price
purchase2_cost = 10 * apple_price + 1 * banana_price

print(f"\n🛒 Purchase verification:")
print(f"   Purchase 1 (2🍎 + 3🍌): €{purchase1_cost:.2f} = €8.00 ✓")
print(f"   Purchase 2 (10🍎 + 1🍌): €{purchase2_cost:.2f} = €13.00 ✓")

# Additional insights
ratio = banana_price / apple_price
print(f"\n📊 Price insights:")
print(f"   • Bananas are {ratio:.2f}× more expensive than apples")
print(f"   • Price difference: €{banana_price - apple_price:.2f} per unit")

if banana_price > apple_price:
    print(f"   • 🍌 Bananas are the premium fruit in this market!")
else:
    print(f"   • 🍎 Apples are more expensive than bananas!")

print(f"\n💡 Key Insight: This simple fruit pricing problem demonstrates the power")
print(f"   of linear algebra - we've transformed word problems into geometric problems!")
````

## 2. The Matrix Equation: Ax = b (The Heart of Linear Algebra)

### 🎯 Rewriting as a Matrix Equation

We transform our system into the fundamental form of linear algebra:

$$A \cdot x = b$$

Where:

$$A = \begin{bmatrix} 2 & 3 \\ 10 & 1 \end{bmatrix}, \quad x = \begin{bmatrix} a \\ b \end{bmatrix}, \quad b = \begin{bmatrix} 8 \\ 13 \end{bmatrix}$$

### 🔍 What Each Component Represents

| **Component** | **Mathematical Role** | **Geometric Interpretation** | **Real-World Meaning** |
|---------------|----------------------|----------------------------|----------------------|
| **Matrix A** | Coefficient container | **Transformation machine** | How ingredients combine |
| **Vector x** | Unknown variables | **Input to transformation** | Quantities we seek |
| **Vector b** | Target values | **Output of transformation** | Observed results |

### 🎨 The Revolutionary Perspective Shift

#### **Traditional View**: "Solve for unknowns"
```
2a + 3b = 8     ← Find a and b that satisfy this
10a + 1b = 13   ← Find a and b that satisfy this too
```

#### **Matrix View**: "What input produces this output?"
```
A × x = b   ← What vector x, when transformed by A, gives b?
```

### 🔧 Understanding Matrix Multiplication

When we compute A × x, we're actually calculating:

$$A \cdot x = \begin{bmatrix} 2 & 3 \\ 10 & 1 \end{bmatrix} \begin{bmatrix} a \\ b \end{bmatrix} = \begin{bmatrix} 2a + 3b \\ 10a + 1b \end{bmatrix}$$

**This reveals the magic**: Matrix multiplication is **linear combination of columns**!

$$A \cdot x = a \cdot \begin{bmatrix} 2 \\ 10 \end{bmatrix} + b \cdot \begin{bmatrix} 3 \\ 1 \end{bmatrix}$$

### 💡 The Geometric Insight

Matrix A defines a **transformation** that:
- Takes the standard basis vectors: $e_1 = [1,0]$, $e_2 = [0,1]$ 
- Maps them to: $A \cdot e_1 = [2,10]$, $A \cdot e_2 = [3,1]$
- **Any vector x gets mapped** according to this transformation

### 🎯 The Core Question

Solving $A \cdot x = b$ asks:

> **"What combination of the transformed basis vectors [2,10] and [3,1] produces the target vector [8,13]?"**

This is the essence of linear algebra:
- **Understanding transformations** of space
- **Finding combinations** that reach specific targets
- **Connecting algebra with geometry**

2. **📊 Data Analysis**:

## 3. Geometric Meaning: Matrix as a Space Transformer

### 🎨 Matrices Transform Space

Every matrix is a **transformation machine** that reshapes coordinate space. To understand what matrix A does, we need to see where it sends the **standard basis vectors**.

### 📐 The Standard Basis Vectors

In 2D, every vector can be written as a combination of:
- $\mathbf{e_1} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ (unit vector along x-axis)
- $\mathbf{e_2} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ (unit vector along y-axis)

**Any vector** $\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix} = x \cdot \mathbf{e_1} + y \cdot \mathbf{e_2}$

### 🔄 What Matrix A Does

Our matrix $A = \begin{bmatrix} 2 & 3 \\ 10 & 1 \end{bmatrix}$ transforms:

- $\mathbf{e_1} = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \rightarrow A \cdot \mathbf{e_1} = \begin{bmatrix} 2 \\ 10 \end{bmatrix}$
- $\mathbf{e_2} = \begin{bmatrix} 0 \\ 1 \end{bmatrix} \rightarrow A \cdot \mathbf{e_2} = \begin{bmatrix} 3 \\ 1 \end{bmatrix}$

### 💡 The Fundamental Insight

**Matrix columns = transformed basis vectors!**

$$A = \begin{bmatrix} 2 & 3 \\ 10 & 1 \end{bmatrix} = \begin{bmatrix} A\mathbf{e_1} & A\mathbf{e_2} \end{bmatrix}$$

### 🎯 Understanding the Transformation

| **Property** | **Original Space** | **After Transformation** |
|--------------|-------------------|-------------------------|
| **Basis 1** | $[1, 0]$ | $[2, 10]$ ← Much longer, steeper |
| **Basis 2** | $[0, 1]$ | $[3, 1]$ ← Longer, shallower |
| **Area** | Unit square = 1 | Parallelogram area = $\det(A) = -28$ |
| **Orientation** | Counter-clockwise | Clockwise (det < 0) |

### 🔍 Geometric Properties

1. **🔄 Linear Transformation**: Lines remain lines, origin stays fixed
2. **📏 Scaling**: Distances change by factors determined by matrix
3. **🌀 Rotation/Shear**: Angles between vectors change
4. **📐 Area Change**: All areas scaled by $|\det(A)|$

### 🎪 The Magic of Linear Combinations

For any vector $\mathbf{x} = \begin{bmatrix} a \\ b \end{bmatrix}$:

$$A \cdot \mathbf{x} = a \cdot (A\mathbf{e_1}) + b \cdot (A\mathbf{e_2}) = a \cdot \begin{bmatrix} 2 \\ 10 \end{bmatrix} + b \cdot \begin{bmatrix} 3 \\ 1 \end{bmatrix}$$

**This means**: To find where A sends any vector, just take the appropriate linear combination of the transformed basis vectors!

### 🎯 Our Solution in Geometric Terms

When we solve $A \cdot \mathbf{x} = \mathbf{b}$, we're asking:

> **"What coefficients $a$ and $b$ make the linear combination $a \cdot [2,10] + b \cdot [3,1]$ equal to $[8,13]$?"**

This is why our solution $a ≈ 1.11, b ≈ 1.93$ works:
- Take 1.11 units of vector $[2,10]$ 
- Add 1.93 units of vector $[3,1]$
- Result: exactly the target $[8,13]$!

Let's visualize this transformation and see the magic happen! ✨


````python
# === COMPREHENSIVE GEOMETRIC VISUALIZATION ===
print("🎨 VISUALIZING MATRIX TRANSFORMATIONS AND SOLUTIONS")
print("="*65)

# === STEP 1: Calculate Transformed Basis Vectors ===
print("📊 STEP 1: Understanding the Transformation")
print("="*45)

# Original basis vectors
e1_original = np.array([1, 0])
e2_original = np.array([0, 1])

# Transformed basis vectors (columns of A)
e1_transformed = A @ e1_original  # First column of A
e2_transformed = A @ e2_original  # Second column of A

print(f"🔍 Original basis vectors:")
print(f"   e₁ = {e1_original} (standard x-axis)")
print(f"   e₂ = {e2_original} (standard y-axis)")

print(f"\n🎯 Transformed by matrix A:")
print(f"   A·e₁ = {e1_transformed} (new direction 1)")
print(f"   A·e₂ = {e2_transformed} (new direction 2)")

# Calculate some geometric properties
angle_between = np.arccos(np.dot(e1_transformed, e2_transformed) / 
                         (np.linalg.norm(e1_transformed) * np.linalg.norm(e2_transformed)))
area_change = abs(np.linalg.det(A))

print(f"\n📐 Geometric properties:")
print(f"   Angle between transformed basis vectors: {np.degrees(angle_between):.1f}°")
print(f"   Area scaling factor: {area_change:.1f}")
print(f"   Determinant: {np.linalg.det(A):.1f} (negative → orientation flipped)")

# === STEP 2: Create Comprehensive Visualization ===
print(f"\n🎨 STEP 2: Creating Visualizations")
print("="*45)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 14))

# === Plot 1: Original vs Transformed Basis Vectors ===
ax1.set_title('Basis Vector Transformation', fontsize=14, fontweight='bold')

# Original basis vectors
ax1.quiver(0, 0, e1_original[0], e1_original[1], angles='xy', scale_units='xy', scale=1, 
           color='lightblue', width=0.01, alpha=0.6, label='Original e₁')
ax1.quiver(0, 0, e2_original[0], e2_original[1], angles='xy', scale_units='xy', scale=1, 
           color='lightcoral', width=0.01, alpha=0.6, label='Original e₂')

# Transformed basis vectors
ax1.quiver(0, 0, e1_transformed[0], e1_transformed[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.008, label=f'A·e₁ = {e1_transformed}')
ax1.quiver(0, 0, e2_transformed[0], e2_transformed[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.008, label=f'A·e₂ = {e2_transformed}')

# Target vector b
ax1.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.01, label=f'Target b = {b}')

ax1.set_xlim(-2, 12)
ax1.set_ylim(-2, 12)
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_aspect('equal')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# === Plot 2: Linear Combination Construction ===
ax2.set_title('Solution as Linear Combination', fontsize=14, fontweight='bold')

# Show step-by-step construction
step1 = apple_price * e1_transformed  # First component
step2 = banana_price * e2_transformed  # Second component
final_result = step1 + step2  # Should equal b

# Plot the construction
ax2.quiver(0, 0, step1[0], step1[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.008, alpha=0.8, label=f'{apple_price:.2f} × A·e₁')
ax2.quiver(step1[0], step1[1], step2[0], step2[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.008, alpha=0.8, label=f'{banana_price:.2f} × A·e₂')

# Show the path taken
ax2.plot([0, step1[0]], [0, step1[1]], 'b-', linewidth=3, alpha=0.6)
ax2.plot([step1[0], final_result[0]], [step1[1], final_result[1]], 'r-', linewidth=3, alpha=0.6)

# Show key points
ax2.plot(0, 0, 'ko', markersize=8, label='Start (0,0)')
ax2.plot(step1[0], step1[1], 'bo', markersize=8, alpha=0.7)
ax2.plot(final_result[0], final_result[1], 'go', markersize=10, label=f'Result {final_result}')

# Target for comparison (using plot instead of quiver to avoid linestyle issues)
ax2.arrow(0, 0, b[0], b[1], head_width=0.3, head_length=0.3, fc='green', ec='green', 
          alpha=0.5, label=f'Target b = {b}')

ax2.set_xlim(-1, 10)
ax2.set_ylim(-1, 12)
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_aspect('equal')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Add annotation arrows
ax2.annotate('Step 1: Scale A·e₁', xy=(step1[0]/2, step1[1]/2), xytext=(1, 8),
             arrowprops=dict(arrowstyle='->', color='blue', lw=2),
             fontsize=10, color='blue', fontweight='bold')
ax2.annotate('Step 2: Add scaled A·e₂', xy=(step1[0] + step2[0]/2, step1[1] + step2[1]/2), 
             xytext=(6, 2), arrowprops=dict(arrowstyle='->', color='red', lw=2),
             fontsize=10, color='red', fontweight='bold')

# === Plot 3: Grid Transformation Visualization ===
ax3.set_title('How Matrix A Transforms Space', fontsize=14, fontweight='bold')

# Create a grid of points to show transformation
x_grid, y_grid = np.meshgrid(np.linspace(-0.5, 1.5, 5), np.linspace(-0.5, 1.5, 5))
original_points = np.vstack([x_grid.ravel(), y_grid.ravel()])
transformed_points = A @ original_points

# Plot original grid
ax3.scatter(original_points[0], original_points[1], c='lightblue', s=30, alpha=0.6, label='Original grid')

# Plot transformed grid  
ax3.scatter(transformed_points[0], transformed_points[1], c='red', s=30, label='Transformed grid')

# Connect corresponding points with arrows
for i in range(0, len(original_points[0]), 3):  # Show every 3rd point to avoid clutter
    ax3.annotate('', xy=transformed_points[:, i], xytext=original_points[:, i],
                arrowprops=dict(arrowstyle='->', color='gray', alpha=0.5))

# Highlight basis vectors transformation
ax3.quiver(0, 0, e1_original[0], e1_original[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.01, alpha=0.6)
ax3.quiver(0, 0, e2_original[0], e2_original[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.01, alpha=0.6)
ax3.quiver(0, 0, e1_transformed[0], e1_transformed[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.008)
ax3.quiver(0, 0, e2_transformed[0], e2_transformed[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.008)

ax3.set_xlim(-1, 15)
ax3.set_ylim(-1, 15)
ax3.grid(True, alpha=0.3)
ax3.legend()
ax3.set_xlabel('x')
ax3.set_ylabel('y')

# === Plot 4: Solution Verification ===
ax4.set_title('Verification: A × solution = target', fontsize=14, fontweight='bold')

# Show the complete solution process
solution_vector = np.array([apple_price, banana_price])
result_vector = A @ solution_vector

# Plot input and output
ax4.quiver(0, 0, solution_vector[0], solution_vector[1], angles='xy', scale_units='xy', scale=1, 
           color='purple', width=0.01, label=f'Solution x = [{solution_vector[0]:.2f}, {solution_vector[1]:.2f}]')
ax4.quiver(0, 0, result_vector[0], result_vector[1], angles='xy', scale_units='xy', scale=1, 
           color='orange', width=0.01, label=f'A×x = {result_vector}')

# Use arrow for target to avoid linestyle issues
ax4.arrow(0, 0, b[0], b[1], head_width=0.3, head_length=0.3, fc='green', ec='green', 
          alpha=0.7, label=f'Target b = {b}')

# Add text showing the verification
verification_text = f'Verification:\nA × x = {result_vector}\nTarget b = {b}\nMatch: {np.allclose(result_vector, b)}'
ax4.text(0.02, 0.98, verification_text, transform=ax4.transAxes, 
         bbox=dict(boxstyle="round,pad=0.5", facecolor="yellow", alpha=0.8),
         fontsize=10, verticalalignment='top')

ax4.set_xlim(-0.5, 15)
ax4.set_ylim(-0.5, 15)
ax4.grid(True, alpha=0.3)
ax4.legend()
ax4.set_xlabel('x')
ax4.set_ylabel('y')

plt.tight_layout()
plt.show()

# === STEP 3: Numerical Summary ===
print(f"\n📊 STEP 3: Numerical Summary")
print("="*45)

print(f"🎯 Solution interpretation:")
print(f"   • Apple price (a): €{apple_price:.4f}")
print(f"   • Banana price (b): €{banana_price:.4f}")

print(f"\n🔍 Vector combination:")
print(f"   {apple_price:.3f} × {e1_transformed} + {banana_price:.3f} × {e2_transformed}")
print(f"   = {step1} + {step2}")
print(f"   = {final_result}")
print(f"   = Target {b} ✓")

print(f"\n💡 Geometric insight:")
print(f"   The solution vector [{apple_price:.3f}, {banana_price:.3f}] tells us exactly")
print(f"   how much of each transformed basis vector we need to reach our target!")

error = np.linalg.norm(final_result - b)
print(f"\n✅ Verification: ||A×x - b|| = {error:.2e} (should be ≈ 0)")
````

## 4. The Solution: Finding the Perfect Linear Combination

### 🎯 The Fundamental Question

We're solving: $$A \cdot x = b$$

Which expands to: $$x_1 \cdot (A \cdot e_1) + x_2 \cdot (A \cdot e_2) = b$$

**In our case**: $$a \cdot \begin{bmatrix} 2 \\ 10 \end{bmatrix} + b \cdot \begin{bmatrix} 3 \\ 1 \end{bmatrix} = \begin{bmatrix} 8 \\ 13 \end{bmatrix}$$

### 🔍 What This Means Geometrically

We're asking: **"What linear combination of the transformed basis vectors reaches our target?"**

| **Component** | **Geometric Meaning** | **Our Problem** |
|---------------|----------------------|-----------------|
| **Transformed basis vectors** | New coordinate directions | $[2,10]$ and $[3,1]$ |
| **Coefficients** | How much to scale each direction | $a$ and $b$ (what we solve for) |
| **Target vector** | Destination point | $[8,13]$ |
| **Linear combination** | Path to reach the target | Weighted sum of directions |

### 💡 The Beautiful Insight

**Matrix-vector multiplication is linear combination of columns!**

When we compute $A \cdot x$, we're really computing:
$$A \cdot x = x_1 \cdot \text{(column 1 of A)} + x_2 \cdot \text{(column 2 of A)} + \cdots$$

### 🎨 Step-by-Step Construction

Our solution process:

1. **🎯 Start at origin**: $(0, 0)$
2. **➡️ Step 1**: Move $a = 1.11$ units along vector $[2, 10]$
   - Reach position: $1.11 \times [2, 10] = [2.21, 11.07]$
3. **➡️ Step 2**: From there, move $b = 1.93$ units along vector $[3, 1]$
   - Add displacement: $1.93 \times [3, 1] = [5.79, 1.93]$
4. **🎯 Final position**: $[2.21, 11.07] + [5.79, 1.93] = [8, 13]$ ✓

### 🔧 Why This Works: Linear Combinations

The power of linear combinations:
- **🌈 Span**: The set of all possible linear combinations of vectors
- **🎯 Reachability**: We can reach any point in the span
- **🔀 Uniqueness**: If vectors are linearly independent, there's exactly one way

### 🚀 Connecting to Real-World Applications

This same principle appears everywhere:

#### **🤖 Machine Learning**
```
prediction = w₁ × feature₁ + w₂ × feature₂ + ... + wₙ × featureₙ
```
- Features are directions in high-dimensional space
- Weights are coefficients (what we learn)
- Prediction is the linear combination result

#### **📊 Principal Component Analysis**
```
data_point = c₁ × PC₁ + c₂ × PC₂ + ... + cₖ × PCₖ
```
- Principal components are new basis directions
- Coefficients tell us position in new coordinate system

#### **🎮 Computer Graphics**
```
transformed_vertex = matrix × original_vertex
```
- Matrix columns define how to transform 3D space
- Result is new position after transformation

### 💭 The Conceptual Breakthrough

> **Traditional thinking**: "Solve equations to find unknowns"
> 
> **Vector thinking**: "Find the recipe to combine ingredients and reach a target"

This shift in perspective is what makes linear algebra so powerful and applicable across virtually every quantitative field!

````python
# === COMPREHENSIVE SOLUTION VERIFICATION ===
print("✅ VERIFYING OUR SOLUTION: Multiple Perspectives")
print("="*60)

# === STEP 1: Direct Reconstruction ===
print("📊 STEP 1: Linear Combination Reconstruction")
print("="*50)

# Calculate the linear combination step by step
component1 = apple_price * e1_transformed
component2 = banana_price * e2_transformed
reconstructed_b = component1 + component2

print(f"🔍 Step-by-step reconstruction:")
print(f"   Apple coefficient × A·e₁ = {apple_price:.4f} × {e1_transformed} = {component1}")
print(f"   Banana coefficient × A·e₂ = {banana_price:.4f} × {e2_transformed} = {component2}")
print(f"   Sum (reconstructed b) = {component1} + {component2} = {reconstructed_b}")
print(f"   Target b = {b}")

# Check if they match
match = np.allclose(reconstructed_b, b)
error = np.linalg.norm(reconstructed_b - b)
print(f"   ✅ Perfect match: {match}")
print(f"   📊 Numerical error: {error:.2e}")

# === STEP 2: Matrix Multiplication Verification ===
print(f"\n📊 STEP 2: Matrix Multiplication Check")
print("="*50)

solution_vector = np.array([apple_price, banana_price])
matrix_result = A @ solution_vector

print(f"🔢 Matrix multiplication: A × x")
print(f"   Matrix A = \n{A}")
print(f"   Solution x = {solution_vector}")
print(f"   A × x = {matrix_result}")
print(f"   Target b = {b}")
print(f"   ✅ Match: {np.allclose(matrix_result, b)}")

# === STEP 3: Original Equations Check ===
print(f"\n📊 STEP 3: Original Equations Verification")
print("="*50)

# Check both original equations
eq1_left = 2 * apple_price + 3 * banana_price
eq2_left = 10 * apple_price + 1 * banana_price

print(f"🍎🍌 Substituting back into original equations:")
print(f"   Equation 1: 2 × {apple_price:.4f} + 3 × {banana_price:.4f} = {eq1_left:.6f}")
print(f"   Should equal: 8")
print(f"   ✅ Difference: {abs(eq1_left - 8):.2e}")

print(f"\n   Equation 2: 10 × {apple_price:.4f} + 1 × {banana_price:.4f} = {eq2_left:.6f}")
print(f"   Should equal: 13")
print(f"   ✅ Difference: {abs(eq2_left - 13):.2e}")

# === STEP 4: Geometric Verification ===
print(f"\n📊 STEP 4: Geometric Analysis")
print("="*50)

# Calculate some geometric properties
v1_length = np.linalg.norm(e1_transformed)
v2_length = np.linalg.norm(e2_transformed)
target_length = np.linalg.norm(b)
result_length = np.linalg.norm(reconstructed_b)

print(f"📐 Vector lengths:")
print(f"   ||A·e₁|| = {v1_length:.4f}")
print(f"   ||A·e₂|| = {v2_length:.4f}")
print(f"   ||target b|| = {target_length:.4f}")
print(f"   ||reconstructed|| = {result_length:.4f}")

# Check angles
angle_to_v1 = np.degrees(np.arccos(np.dot(b, e1_transformed) / (target_length * v1_length)))
angle_to_v2 = np.degrees(np.arccos(np.dot(b, e2_transformed) / (target_length * v2_length)))

print(f"\n📐 Angular relationships:")
print(f"   Angle between target and A·e₁: {angle_to_v1:.1f}°")
print(f"   Angle between target and A·e₂: {angle_to_v2:.1f}°")

# === STEP 5: Economic Interpretation ===
print(f"\n📊 STEP 5: Economic Verification")
print("="*50)

print(f"💰 Let's verify with the original purchases:")

# Purchase 1: 2 apples + 3 bananas should cost €8
purchase1_cost = 2 * apple_price + 3 * banana_price
print(f"   Purchase 1: 2 apples + 3 bananas")
print(f"   Cost: 2 × €{apple_price:.2f} + 3 × €{banana_price:.2f} = €{purchase1_cost:.2f}")
print(f"   Expected: €8.00")
print(f"   ✅ Difference: €{abs(purchase1_cost - 8):.4f}")

# Purchase 2: 10 apples + 1 banana should cost €13
purchase2_cost = 10 * apple_price + 1 * banana_price
print(f"\n   Purchase 2: 10 apples + 1 banana")
print(f"   Cost: 10 × €{apple_price:.2f} + 1 × €{banana_price:.2f} = €{purchase2_cost:.2f}")
print(f"   Expected: €13.00")
print(f"   ✅ Difference: €{abs(purchase2_cost - 13):.4f}")

# === STEP 6: Alternative Purchase Predictions ===
print(f"\n📊 STEP 6: Making New Predictions")
print("="*50)

print(f"🔮 Now that we know the prices, let's predict other combinations:")

# Predict some new combinations
combinations = [
    (1, 1, "1 apple + 1 banana"),
    (5, 2, "5 apples + 2 bananas"), 
    (0, 4, "4 bananas only"),
    (3, 0, "3 apples only")
]

for apples, bananas, description in combinations:
    predicted_cost = apples * apple_price + bananas * banana_price
    print(f"   {description}: €{predicted_cost:.2f}")

# === FINAL SUMMARY ===
print(f"\n🎯 VERIFICATION SUMMARY")
print("="*50)

all_checks_pass = (
    np.allclose(reconstructed_b, b) and
    np.allclose(matrix_result, b) and 
    abs(eq1_left - 8) < 1e-10 and
    abs(eq2_left - 13) < 1e-10
)

print(f"✅ Linear combination reconstruction: PASS")
print(f"✅ Matrix multiplication check: PASS")
print(f"✅ Original equations verification: PASS")
print(f"✅ Economic interpretation: PASS")
print(f"\n🎉 Overall verification: {'PERFECT SUCCESS' if all_checks_pass else 'FAILED'}")

print(f"\n💡 Key Insight: The solution works because our matrix A has full rank")
print(f"   (linearly independent columns), guaranteeing a unique solution exists!")

print(f"\n🚀 This verification demonstrates the power of the matrix-vector perspective:")
print(f"   We've connected algebra, geometry, and real-world meaning in one coherent framework!")
````

````python
# === INTERACTIVE EXERCISES: APPLY YOUR KNOWLEDGE ===
print("🎯 INTERACTIVE EXERCISES: Test Your Understanding!")
print("="*60)

def analyze_matrix_system(A, b, name="System"):
    """
    Analyze any 2x2 matrix system using our new perspective
    """
    print(f"\n📊 Analyzing {name}")
    print("="*40)
    
    # System info
    print(f"Matrix A:\n{A}")
    print(f"Target b: {b}")
    
    # Check if solvable
    det_A = np.linalg.det(A)
    print(f"\nDeterminant: {det_A:.4f}")
    
    if abs(det_A) < 1e-10:
        print("❌ System is singular (det ≈ 0) - no unique solution")
        return None
    
    # Solve
    solution = np.linalg.solve(A, b)
    print(f"✅ Solution: {solution}")
    
    # Geometric analysis
    e1_trans = A @ np.array([1, 0])
    e2_trans = A @ np.array([0, 1])
    
    print(f"\n🎨 Geometric interpretation:")
    print(f"   Transformed basis vectors:")
    print(f"   A·e₁ = {e1_trans}")
    print(f"   A·e₂ = {e2_trans}")
    
    # Linear combination
    reconstruction = solution[0] * e1_trans + solution[1] * e2_trans
    print(f"\n🔍 Linear combination:")
    print(f"   {solution[0]:.3f} × {e1_trans} + {solution[1]:.3f} × {e2_trans}")
    print(f"   = {reconstruction}")
    print(f"   Target: {b}")
    print(f"   ✅ Match: {np.allclose(reconstruction, b)}")
    
    return solution

# === EXERCISE 1: Different Fruit Market ===
print("🍊🍌 EXERCISE 1: Orange and Banana Market")
A1 = np.array([[1, 2],
               [3, 1]])
b1 = np.array([5, 8])
print("Scenario: 1 orange + 2 bananas = €5, 3 oranges + 1 banana = €8")
sol1 = analyze_matrix_system(A1, b1, "Orange-Banana Market")

# === EXERCISE 2: Scaling Transformation ===
print("\n🔄 EXERCISE 2: Simple Scaling")
A2 = np.array([[2, 0],
               [0, 3]])
b2 = np.array([4, 9])
print("Scenario: Pure scaling transformation")
sol2 = analyze_matrix_system(A2, b2, "Scaling Transformation")

# === EXERCISE 3: Rotation-like Transformation ===
print("\n🌀 EXERCISE 3: Rotation-like System")
A3 = np.array([[1, -1],
               [1, 1]])
b3 = np.array([2, 4])
print("Scenario: Rotation-like transformation")
sol3 = analyze_matrix_system(A3, b3, "Rotation-like System")

# === VISUALIZATION OF ALL EXERCISES ===
print(f"\n🎨 VISUALIZING ALL EXERCISES")
print("="*40)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
systems = [(A1, b1, sol1, "Orange-Banana"), (A2, b2, sol2, "Scaling"), (A3, b3, sol3, "Rotation-like")]

for i, (A, b, sol, name) in enumerate(systems):
    ax = axes[i]
    
    # Transformed basis vectors
    e1_trans = A @ np.array([1, 0])
    e2_trans = A @ np.array([0, 1])
    
    # Plot transformation
    ax.quiver(0, 0, e1_trans[0], e1_trans[1], angles='xy', scale_units='xy', scale=1, 
              color='blue', width=0.008, label=f'A·e₁ = {e1_trans}')
    ax.quiver(0, 0, e2_trans[0], e2_trans[1], angles='xy', scale_units='xy', scale=1, 
              color='red', width=0.008, label=f'A·e₂ = {e2_trans}')
    
    # Target and solution
    ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, 
              color='green', width=0.01, label=f'Target {b}')
    
    # Show linear combination construction
    if sol is not None:
        step1 = sol[0] * e1_trans
        step2 = sol[1] * e2_trans
        ax.plot([0, step1[0]], [0, step1[1]], 'b-', linewidth=2, alpha=0.6)
        ax.plot([step1[0], step1[0] + step2[0]], [step1[1], step1[1] + step2[1]], 'r-', linewidth=2, alpha=0.6)
        ax.plot(step1[0], step1[1], 'bo', markersize=6)
        ax.plot(step1[0] + step2[0], step1[1] + step2[1], 'go', markersize=8)
    
    ax.set_title(f'{name}\nSolution: {sol}', fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.legend()
    ax.set_aspect('equal')
    
    # Set reasonable limits
    all_points = np.column_stack([e1_trans, e2_trans, b])
    margin = 1
    ax.set_xlim(np.min(all_points[0]) - margin, np.max(all_points[0]) + margin)
    ax.set_ylim(np.min(all_points[1]) - margin, np.max(all_points[1]) + margin)

plt.tight_layout()
plt.show()

# === CHALLENGE PROBLEMS ===
print(f"\n🏆 CHALLENGE YOURSELF!")
print("="*40)

print("""
Try these on your own:

1. 🎯 Create a system where the solution is [1, 1]:
   What matrix A and target b would give this solution?

2. 🔄 Design a transformation that:
   - Doubles the x-coordinate
   - Rotates 90° counter-clockwise
   What would this matrix look like?

3. 📊 Real-world problem:
   If 2 hours of work + 3 hours of study = 50 points
   And 4 hours of work + 1 hour of study = 46 points
   What's the value of each hour?

4. 🎨 Geometric challenge:
   What happens when det(A) = 0? 
   Create an example and visualize it!

Use the analyze_matrix_system() function above to check your answers!
""")

# === NEXT STEPS ===
print(f"\n🚀 NEXT STEPS IN YOUR LINEAR ALGEBRA JOURNEY")
print("="*60)

print("""
Now that you understand matrix-vector thinking, explore:

📚 Theory:
   • Eigenvalues and eigenvectors
   • Matrix decompositions (LU, QR, SVD)
   • Vector spaces and linear independence

🤖 Applications:
   • Linear regression and machine learning
   • Principal Component Analysis (PCA)
   • Computer graphics and 3D transformations

🔧 Practice:
   • Solve 3×3 and larger systems
   • Implement matrix operations from scratch
   • Visualize transformations in 3D

💡 Remember: Every advanced concept builds on this foundation of
   "matrices as transformations" and "solutions as linear combinations"!
""")

print("\n🎉 Congratulations on mastering the matrix-vector perspective! 🎉")
````
