# Histogram

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df.drop("Total", axis = 1, inplace = True)
```

```python
df.head()
```

```python
years = list(map(str, range(1980, 2013)))
years
```

```python
df["Total"] = df[years].sum(axis = 1)
```

```python
df.set_index("Country", inplace = True)
```

```python
df.head()
```

```python
df["2013"].plot(kind = "hist")

plt.title("Countries in 2013")
plt.xlabel("Number of histograms")
plt.ylabel("Number of Countries")
```

```python
# adjust bins 

count, bin_edges = np.histogram(df["2013"])
```

```python
df["2013"].plot(kind = "hist", xticks = bin_edges, figsize = (14,8))

plt.title("Countries in 2013")
plt.xlabel("Number of histograms")
plt.ylabel("Number of Countries")
```
