```python
import requests
from bs4 import BeautifulSoup

# URL der Wikipedia-Seite
url = "https://de.wikipedia.org/wiki/Python_(Programmiersprache)"

# HTTP GET Anfrage an die URL
response = requests.get(url)

# Überprüfen, ob die Anfrage erfolgreich war (Statuscode 200)
if response.status_code == 200:
    # Inhalt der Seite parsen
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Beispiel: Den ersten Absatz im Artikel scrapen
    first_paragraph = soup.find('p').text
    print("Erster Absatz des Artikels:")
    print(first_paragraph)
    
    # Beispiel: Alle Überschriften der ersten Ebene scrapen
    headers = soup.find_all('h1')
    print("\nÜberschriften der ersten Ebene:")
    for header in headers:
        print(header.text.strip())

    # Beispiel: Alle Links im Artikel scrapen
    links = soup.find_all('a', href=True)
    print("\nAlle Links im Artikel:")
    for link in links:
        print(link['href'])

else:
    print("Fehler beim Abrufen der Seite:", response.status_code)
```
